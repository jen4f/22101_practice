2\. Given that you have a string of comma separated values, like "1,22,333,4,5". How would you find the positions of the commas in the string?


```python
# pseudo code
"""
prompt user for a string of comma separated values, like 1,22,333,4,5.
to find the comma positions, use a loop that is equal to range(len(str))
initiate a list to record the positions of ",".
for position in range(len(str))
    if str[position] == ","
        list.append(position).
    else, do nothing.
the list records all the positions of ",".
"""
# code
usr_csv = input("Please input a string of comma separated values, like 1,22,333,4,5: ")
# initiate empty list to track "," positions:
comma_pos = []
for position in range(len(usr_csv)):
    if usr_csv[position] == ",":
        comma_pos.append(position)
    else:
        pass
print(f"The positions of the commas in the string are {comma_pos}.")
```

3\. In the files data1-4.gb find all titles and authors using stateful parsing. Put every author/title record (i.e. all authors/title in one string) in one list. Display all neatly with title on one line, authors on the next and a separator line.

```python
# pseudo code
"""
red flag is when tile line has ended, i.e. JOURNAL starts
we want the contents that comes after AUTHOR, and before JOURNAL, putting every author/title record in one string, in one list.
green flag is when AUTHORS line has started
* remember str.startswith method is very helpful.
had to place the order this way:
# green flag
# red flag
# data gathering.

Then, needed to use.strip() + " " while collecting the contents of each lines, so that it display correctly later.
Use .replace() twice to later correctly splice the string into a single list that to be displayed :).

Use a more specific method instead of the a,b = b,a to display the result correctly.

"""
# code

# prompt user to enter the file name.
def main():
    # using one file first, will change into all files later.
    data_files = ["data1.gb", "data2.gb", "data3.gb", "data4.gb"]
    for data_file in data_files:
        display_auth_title(data_file)

def display_auth_title(file):
    with open (f"{file}", "r") as infile:
    # initiat stateful parsing with an empty author_title string, with the flag set to false.
        (auth_title_v1, auth_title_flag) = ("", False)
        # The below flag positons allows the words "AUTHORS", "TITLE" to remain in the final string.
        for line in infile:
            if line.strip().startswith("AUTHORS"):    # green flag
                auth_title_flag = True
            if line.strip().startswith("JOURNAL"):    # red flag
                auth_title_flag = False
            if auth_title_flag:
                #.strip()+" " makes sure places that needed a space have it inplace correctly
                # this works better than .replace("\n". " ")
                auth_title_v1 += line[:].strip()+" "    
        if auth_title_flag:
            raise ValueError("Format is not right. Can't trust result.")
        if auth_title_v1 == "":
            raise ValueError("No data found. Can't trust result.")
    # splite the string using two delimiters 
    auth_title_v2 = auth_title_v1.replace("AUTHORS   ", "%").replace("TITLE     ", "%")
    auth_title_list1 = auth_title_v2.split("%")
    auth_title_list1.pop(0)
    # get rid of the empty space item at the start:
    # the auth_title_list1 items are authors, title, authors, title
    # first switch the positions of the items
    # to have title infront of author.
    # method below to switch positions worked best. The usual a,b = b,a ended up with incorrect items at the start or end.
    new_positions = []
    for position in range(len(auth_title_list1)):
        if position % 2 == 1:
            new_positions.append(position)
            new_positions.append(position-1)

    for position in new_positions:
        print(auth_title_list1[position])       # prints title then author
        if position % 2 == 0:                   # if we are at the end of an author
            print()                             # prints separator line


main()
```

```python
# pseudo code
"""
The code below is copied from ex4
create new function that extracts the position of the coding sequence.
edit the user prompt so they can ask for the CDS positions.
"""

# code
def main():
    usr_filename = input("Please enter a file Genbank entry file, such as one of data1-4.gb: ")
    find_keyword(usr_filename)




def find_keyword(filename):
# make a loop so usr can keep prompting and we can keep displaying
    while True:
        search_word = input("Specify what you want to know by typing one of the following:\naccession number, definition, organism, medline, translated, DNA, CDS positions (STOP to quit search): ")
        if search_word == "medline":
            find_medline(filename)
        elif search_word in ("accession number", "definition", "organism"):
            find_an_df_og(filename, search_word)    # very important to include "search_word" here.
        elif search_word == "translated":
            find_translated(filename)
        elif search_word == "DNA":
            find_DNA(filename)
        elif search_word == "CDS positions":
            find_CDS(filename)
        elif search_word == "STOP":
            break
        else:
            print("I don't think we have what you typed, let's try again.")
            pass



def find_CDS(filename):
    with open(f"{filename}", "r") as infile:
        (result, cds_flag) = ("", False)
        for line in infile:
            line = line.strip()
            if line.startswith("CDS             join("):    # green flag
                cds_flag = True
            if cds_flag:
                result += line
            if line.endswith(")"):                          # red flag
                cds_flag = False
        if cds_flag:
            raise ValueError("Format is not right. Can't trust result.")
        if result == "":
            raise ValueError("No data found. Can't trust result.")
    print("\nBelow are the CDS positions: ")
    for char in result:
        if not char.isalpha():
            char = char.strip("(").strip(")").strip()
            print(char, end = "")
    print()
    print()

def find_DNA(filename):
    with open(f"{filename}", "r") as infile:
        (result, dna_flag) = ("", False)
        for line in infile:
            line = line.strip()
            if line.startswith("//"):        # red flag
                dna_flag = False
            if dna_flag:
                result += ''.join(line.split()[1:])
            if line.startswith("ORIGIN"):    # green flag
                dna_flag = True
        if dna_flag:
            raise ValueError("Format is not right. Can't trust result.")
        if result == "":
            raise ValueError("No data found. Can't trust result.")
    print(f"\nThe DNA sequence of the translated gene is:\n{result}\n")

def find_translated(filename):
    with open(f"{filename}", "r") as infile:
        (result, tran_flag) = ("", False)
        for line in infile:
            line = line.strip()
            # this is the red line
            # put the red line at the top so it is not included in the result :)
            # this way the results stop at the correct place.
            if line.startswith("exon"):
                tran_flag = False
            if tran_flag:
                # carful here, the keyword in the file is translation, not translated :).
                result += line.strip("/translation=").strip('"').strip()
            # this is the green line
            # carful here with, remember the correct word is translation.
            if line.startswith("/translation="):
                tran_flag = True
        if tran_flag:
            raise ValueError("Format is not right. Can't trust result.")
        if result == "":
            raise ValueError("No data found. Can't trust result.")
    print(f"\nThe DNA sequence of the translated gene is:\n{result}\n")


def find_medline(filename):
    medlines = []
    with open(f"{filename}", "r") as infile:
        for line in infile:
            line = line.strip()
            if line.startswith("MEDLINE"):
                medlines.append(line.strip("MEDLINE").strip())
            else:
                pass
    if len(medlines) > 1:
        are = "are"
    else:
        are = "is"
    print(f"The medlines {are}: {medlines}")


def find_an_df_og(filename, search_word):
    with open (f"{filename}", "r") as infile:
        for line in infile:
            # line.strip() will help with key word search like ORGANISM line, which starts with two spaces.
            line = line.strip()
            # the find start parts can be made into a smaller programme, but not that fit for what ex1 asks for.
            # the way to do this is to ask the user for a string, that equals to "ACCESSION", "DEFINITION", ...etc
            # search_word.split()[0] is useful for cases when we write accession number.
            if line.startswith(f"{search_word.split()[0].upper()}"):
                result = line.strip(f"{search_word.split()[0].upper()}").strip()
                print(f"The {search_word} is {result}.")
            elif len(search_word) == 0:
                print("Empty search term!")
            else:
                # using pass here for now
                pass


main()
```
