4\. Add single number operations LOG and SQRT, which replaces the last number in the stack with the log/square root of the number.

```python
# pseudo code
"""
need to import the math module to do log more easily.
"""

# code

import sys
import math


nums_list = []
# changed operators' names to better differentiate
standard_operators = ("+", "-", "*", "/")
single_number_operators = ("LOG", "SQRT")

while True:
    usr_input = input("Enter a number here (To quit, enter STOP, EXIT, or QUIT): ")
    # check if the input string is a number
    if usr_input.isnumeric():
        # simply appending it to the list
        # best to do int(usr_input) at this step
        nums_list.append(int(usr_input))
    # made a tuple of key words to check here
    elif usr_input in ("STOP", "EXIT", "QUIT"):
        break
        sys.exit(1)
    elif usr_input == "PEEK":
        # displays last number in the stack
        print(nums_list[-1])
    elif usr_input == "SHOW":
        # displays the entire stack, printing a copy just in case.
        print(nums_list[:])
    # this line does not have to be here exactly, but it helps to keep things a bit more tidy.
    elif usr_input in standard_operators:
        first_op, second_op = nums_list[-1], nums_list[-2]
        if usr_input == "+":
            result = first_op + second_op
        elif usr_input == "-":
            result = first_op - second_op
        elif usr_input == "*":
            result = first_op * second_op
        elif usr_input == "/":
            result = first_op / second_op
        else:
            pass
        # append the result to the stack
        nums_list.append(result)
    elif usr_input in single_number_operators:
        if usr_input == "LOG":
            result = math.log(nums_list[-1])
        elif usr_input == "SQRT":
            result = math.sqrt(nums_list[-1])
        # append the result to the stack
        nums_list.append(result)
    else:
        print("Illegal command")
        continue
```

6\. Add the commands: DROP which simply removes the last number on the stack, =, which removes the last number on the stack and displays it, SWITCH, which switches the last two numbers on the stack, and DUP which duplicates the last number on the stack, i.e stack consists of 1, after DUP it consists of 1 1.

```python
# pseudo code:
"""
DROP can be achieved using .pop().
= can be achieved by assigning pop to a variable, or simply displaying .pop()
SWITCH can be achieved using a a,b = b,a assignment
DUP can be achieved by appending the list's [-1] index value to the same list.

"""
# code

import sys
import math


nums_list = []
# changed operators' names to better differentiate
two_number_operators = ("+", "-", "*", "/")
single_number_operators = ("LOG", "SQRT")

while True:
    usr_input = input("Enter a number here (To quit, enter STOP, EXIT, or QUIT): ")
    # check if the input string is a number
    if usr_input.isnumeric():
        # simply appending it to the list
        # best to do int(usr_input) at this step
        nums_list.append(int(usr_input))
    # made a tuple of key words to check here
    elif usr_input in ("STOP", "EXIT", "QUIT"):
        break
        sys.exit(1)
    elif usr_input == "PEEK":
        # displays last number in the stack
        print(nums_list[-1])
    elif usr_input == "SHOW":
        # displays the entire stack, printing a copy just in case.
        print(nums_list[:])
    # this line does not have to be here exactly, but it helps to keep things a bit more tidy.
    elif usr_input in two_number_operators:
        if len(nums_list) >= 2:
            first_op, second_op = nums_list[-1], nums_list[-2]
            if usr_input == "+":
                result = first_op + second_op
            elif usr_input == "-":
                result = first_op - second_op
            elif usr_input == "*":
                result = first_op * second_op
            elif usr_input == "/":
                result = first_op / second_op
            else:
                pass
        else:
            print("Not enough operands on the stack")
        # append the result to the stack
        nums_list.append(result)
    elif usr_input in single_number_operators:
        if len(nums_list) >= 1:
            if usr_input == "LOG":
                result = math.log(nums_list[-1])
            elif usr_input == "SQRT":
                result = math.sqrt(nums_list[-1])
            # append the result to the stack
            nums_list.append(result)
        else:
            print("Not enough operands on the stack")
    elif usr_input == "DROP":
        nums_list.pop()
    elif usr_input == "=":
        print(nums_list.pop())
    elif usr_input == "SWITCH":
        nums_list[-1], nums_list[-2] = nums_list[-2], nums_list[-1]
    elif usr_input == "DUP":
        nums_list.append(nums_list[-1])
    else:
        print("Illegal command")
        continue
```
