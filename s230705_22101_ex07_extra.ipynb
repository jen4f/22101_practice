3. Continuing previous exercise: Now also extract the title (RT lines). Display like title first on one line, followed by authors on the next. Then empty line followed by the next title and authors and so forth, until no more authors/titles.

```python
import sys


def main():
    file_to_work_on = input("Enter file name: ")
    #ra_count(file_to_work_on)
    read_ra(file_to_work_on)
    read_rt(file_to_work_on)
    display_rt_ra(file_to_work_on)


def ra_count(filename):
    try:
        infile = open(f"{filename}", "r")
    except IOError as err:
        print("Can't open file, reason:", str(err))
        sys.exit(1)
    
    (swissp_ra_count, raflag) = (0, False)
    for line in infile:
        if line[:2] != "RA":
            raflag = False
        if line[:2] == "RA":
            raflag = True
        if raflag == True:
            swissp_ra_count += 1
    
    if raflag:
        raise ValueError("Format is not right, can not trust result.")
    
    infile.close()
    print(swissp_ra_count)

# extract all authors into one line.
def read_ra(filename):
    try:
        infile = open(f"{filename}", "r")
    except IOError as err:
        print("Can't open file, reason:", str(err))
        sys.exit(1)
    
    (swissp_ra, raflag) = ("", False)
    for line in infile:
        if line[:2] != "RA":
            raflag = False
        if line[:2] == "RA":
            raflag = True
        if raflag == True:
            # adding the space add the end
            # prevents authers in different lines from
            # merging into the previous line
            swissp_ra += line[2:].strip().replace(",", "") + " "
    # first, all the authors from different papers are put into one big string.
    # by replaying ";" with "\n", we correctly put the authors together
    # i.e. authors of the same paper appear on the same line :)
    swissp_ra = swissp_ra.replace(";", "\n").replace("\n ", "\n")

    if raflag:
        raise ValueError("Format is not right, can not trust result.")
    
    infile.close()
#    print(swissp_ra, end = "") #sanity check
    with open(f"{filename}_ra", "w") as outfile:
        outfile.write(swissp_ra)

# extract all authors into one line.
def read_rt(filename):
    try:
        infile = open(f"{filename}", "r")
    except IOError as err:
        print("Can't open file, reason:", str(err))
        sys.exit(1)
    
    (swissp_rt, rtflag) = ("", False)
    for line in infile:
        if line[:2] != "RT":
            rtflag = False
        if line[:2] == "RT":
            rtflag = True
        if rtflag == True:
            # adding the space add the end
            # prevents authers in different lines from
            # merging into the previous line
            swissp_rt += line[2:].strip().replace(",", "") + " "
    # first, all the titles from different papers are put into one big string.
    # by replaying ";" with "\n", we correctly separate each title.
    swissp_rt = swissp_rt.replace(";", "\n").replace(' "', '"')

    if rtflag:
        raise ValueError("Format is not right, can not trust result.")
    
    infile.close()
#    print()
#    print(swissp_rt, end = "") #sanity check
    with open(f"{filename}_rt", "w") as outfile:
        outfile.write(swissp_rt)

def display_rt_ra(filename):
    try:
        rt_file = open(f"{filename}_rt", "r")
        ra_file = open(f"{filename}_ra", "r")
    except IOError as err:
        print("Can't open file, reason:", str(err))
        
    rt_as_list = rt_file.readlines()
    ra_as_list = ra_file.readlines()
    ra_file.close()
    rt_file.close()
    rt_ra = ""
    for i in range(len(ra_as_list)):    # len(rt_as_list) is 1 longer than that of ra.
        rt_ra += rt_as_list[i] + ra_as_list[i] + "\n"
    print(rt_ra, end = "")


main()
```
