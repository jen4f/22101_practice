1\. Ask for some innput - it could be a word or a number or a sentence. Then ask for a file name - any file will do, but mixedlines.txt or ex5.acc are good. Now read the file and determine if one (or more) of the lines in the file matches your input precisely. Display "Match" or "No match" accordingly. Hint: A line from a file has a newline at the end, but using the input function does not result in this newline. 

```python
find_this = input("Enter a word, a number, or a sentence: ")
file_name = input("Enter a file name: ")

with open(f"{file_name}", "r") as infile:
    for line in infile:
        # remove the new line at the end of each line
        # any empty lines orignially in the file will still be there
        line = line.strip()
        if find_this in line:
            print("Match")
        else:
            print("No match")
```

3\. Ask for 2 integers. Now print the first 20 numbers of the [Fibonacci sequence](https://en.wikipedia.org/wiki/Fibonacci_number) given by those numbers. If the input is 1 & 3, the output would be 1, 3, 4, 6, 11, 18, 29, 47, etc.. The next number is the sum of the 2 previous.

```python
# initiate an empty list
int_list = []
# prompts user for the first number
int_list.append(int(input("Enter the first integer: ")))
# prompts user for the second number
int_list.append(int(input("Enter the second integer: ")))

# set the initial index to the second number,
# i.e. the latest number the user entered
i = 1
# 21 - 2 = 19
# we want to get 20 numbers in total
# so it would need to reach i < 21.
# we already have 2 inputs, so we only need to each 18 numbers
# therefore here we write i < 19, to get a total of 20 numbers
# in our output
while i < 19:
    # from the second number the user entered
    # up to the final index
    # we sum the values of what is in the current index
    # with the value of what is in the previous index
    # i.e. The next number is the sum of the 2 previous.
    int_list.append(int_list[i] + int_list[i-1])
    # we increase the index number by one each time
    i += 1
    
for integer in int_list:
    # this makes is to that no comma appears after the final value
    # instead we print a new line
    if int_list.index(integer) == len(int_list) - 1:
        print(integer, end = "\n")
    # this puts a comma after each value
    else:
        print(integer, end = ", ")

print(f"There are {len(int_list)} integers in total.")
```
