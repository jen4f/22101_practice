1\. Write a program that counts the number of negative numbers in ex1.dat file. Display the result.


```python
# pseudocode
"""
convert all "\t" to "\n"
write this out into a new file

initiate counter variable set to 0.

open the file
    for line in the file
        if value of line is below 0
            increse counter variable value by one.

display counter variable value
"""
# code
infile = open("ex1.dat", "r")

content = ""
for line in infile:
    for char in line:
        if char == "\t":
            content += "\n"
        else:
            content += char
infile.close()

# open output file for writing
outfile = open("ex1_no_tabs.dat", "w")
outfile.write(content)
outfile.close()

# count the number of negative values in the new file
counter = 0
with open("ex1_no_tabs.dat", "r") as infile:
    for line in infile:
        line = float(line.strip())
        if line > 0:
            counter += 1
        else:
            pass
print(counter)
```

2\. Write a programme that converts temperatures from Fahrenheit to Celsius or visa versa given input like "36F" or "15C" ( F = (C * 9/5) + 32 ).

```python
# pseudo code
"""
prompt user to enter a temperature
check if the character at the last index of the input is F or C
if that value is C, execute this formula to return the temp in C
    F = (C * (9/5)) + 32
if that value is F, exeecute this formulat to return the temp in F
    C = (F - 32) * (5/9)
"""
# code

temp = input("Enter a temperature such as '36F' or '15C': ")
if temp[-1] == "F":
    temp_F = int(temp[:-1])
    temp_C = (temp_F - 32) * (5/9)
    temp_C = format(temp_C, ".3f")
    print(f"{temp_C}C")
elif temp[-1] == "C":
    temp_C = int(temp[:-1])
    temp_F = (temp_C * (9/5)) + 32
    temp_F = format(temp_F, ".3f")
    print(f"{temp_F}F")
else:
    pass
```

4\. Now you must analyze the AT/GC content of the DNA in the file dna.dat. You must count all A, T, C and G, and display the result.

```python
# pseudo code
"""
initiate an empty dictionary
keys will the A, T, C and G.
values will be their counts

open the dna.dat file
for line in the file
    alter line so there are no \n
    for char in the line
        if the character is A
            append 1 to A
        and so on for T, C, G.

calculate the percentage AT content
and the percentage GC content
print these.
"""
# code
counts = {}

with open("dna.dat", "r") as infile:
    for line in infile:
        line = line.strip()
        for char in line:
            if char not in counts:
                counts[char] = 0
            counts[char] += 1

GC_sum = counts.get("G") + counts.get("C")
AT_sum = counts.get("A") + counts.get("T")

total_count = GC_sum + AT_sum
GC_percentage = format((GC_sum / total_count) * 100, ".3f")
AT_percentage = format((AT_sum/ total_count) * 100, ".3f")
print(f"The percentage of GC content is {GC_percentage}%.")
print(f"The percentage of AT content is {AT_percentage}%.")
```

```python
# pseudo code
"""
main function
call function to create file without header line, and file of the header alone 
call function to create a complementary string 
call function to create a reverse complementary string
call function to make the r.c. string into the pretty format, and append it to the header only file.

define function to remove header line
define function to create a complementary string
define function to create a reverse complementary string
define funciton to make the r.c. string pretty 
define a function that writes the correct header with the r.c. string into an output file.

call the main function

"""
# code
import os


def main():
    inputfile = input("Enter fasta file name: ")
    put_header_away(inputfile)
    complementary = make_complement(inputfile)
    reverse = reverse_strand(complementary)
    r_c_pretty = make_pretty(reverse)
    make_file(r_c_pretty, inputfile)


def put_header_away(fastafile):
    # create two seperate file, one of the header and one without the header.
    os.system(f"tail -n +2 {fastafile} > no_header_{fastafile}")
    os.system(f"head -n 1 {fastafile} > header_only_{fastafile}")


def make_complement(file):
    comp = {"A" : "T", "T" : "A", "C" : "G", "G" : "C"}
    comp_strand = ""
    with open(f"no_header_{file}", "r") as infile:
        for line in infile:
            for char in line.strip():
                comp_strand += comp.get(char)
    return(comp_strand)    # return this for the next function


def reverse_strand(input_string):
    index_value = -1
    reverse_comp_strand = ""
    while len(reverse_comp_strand) != len(input_string):
        reverse_comp_strand += input_string[index_value]
        index_value -= 1
    return(reverse_comp_strand)    # returns this for the next function


def make_pretty(input_string):
    index_value = 0
    reverse_comp_pretty = ""
    # expected filnal length = original length + number
    new_characters = len(input_string)//60
    final_length = len(input_string) + new_characters

    while len(reverse_comp_pretty) != final_length:
        reverse_comp_pretty += input_string[index_value]
        index_value += 1
        if index_value % 60 == 0:
            reverse_comp_pretty += "\n"
        else:
            pass
    return(reverse_comp_pretty)    # returns this for the next function


def make_file(input_string, file):
    # first, edit the identifying line
    with open(f"header_only_{file}", "r") as infile:
        new_header = ""
        for line in infile:
            line = line.strip()
            for char in line:
                new_header += char
        new_header += " ComplementaryStrand\n"
    # second, append the rev.comp. sequence, creating a final out file.
    with open(f"revcomp_{file}", "w") as outfile:
        outfile.write(new_header)
        outfile.write(input_string)


main()
```
