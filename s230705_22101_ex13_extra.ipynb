1\. Compute the [LIX](https://en.wikipedia.org/wiki/Lix_(readability_test)) score of a text. Ask for a file name of a text file. You have to create the file yourself, maybe the abstract of an article or a fairy tale. Count the number of words, the number of long words (6 or more letters) and the number of sentences in the text. Compute the lix score as described on the [wikipedia page](https://en.wikipedia.org/wiki/Lix_(readability_test)).

```python
# pseudo code
"""
1. Ask for a file name of a text file (i.e. biontech_ft_28-03-2023.txt).
2. To track number of words: try read line or try spliting each line and keep appending each item to an initally empty list.
3. To track number of long words: (same loop as 2., add if close that measures len(item).
4. Assuming that sentences will end in (". ", ": ", "! " or "? ") or (".\n", ":\n", "!\n ", "?\n"),  Split on this and count.
    I added a space after the puncuation mark to, at least, differentiate cases of Mr.Smith.
5. Compute the lix score.
"""

# code

# initiate three empty lists.
def main():

    n_words, n_l_words, n_sentences = 0, 0, 0

    with open ("biontech_ft_28-03-2023.txt", "r") as infile:
        for line in infile:
            all_lines = infile.readlines()
    # print(all_lines)    sanity check
    # cannot use count lines to find out number of sentences because sometimes we have "\n".

    for item in all_lines:
        if item == "\n":
            pass
        else:
            n_words += len(item.split())
        for word in item.split():
            if len(word) > 6:
                n_l_words += 1
            else:
                pass
            try:
                if word[-1] == ".":
                    n_sentences += 1
                elif word[-2] == ".":
                    n_sentences += 1
            except:
                IndexError

    print(f"There are {n_words} words in this text.")
    print(f"There are {n_l_words} words with over 6 letters in this text.")
    print(f"There are {n_sentences} sentences in this text.")

    lix_score = get_lix(n_words, n_sentences, n_l_words)
    lix_score = "{:.0f}".format(lix_score)
    print(f"The lix_score of biontech_ft_28-03-2023.txt is {lix_score}")

def get_lix(A, B, C):
# A = number of words, B = number of sentences, c = number of long words
    lix = (A/B) + ((C*100)/A)
    return lix


main()
```

```python
def main():
    while True:
        usr_file_name = input("Please enter the filename, i.e. dna7.fsa (STOP to exit): ")
        if usr_file_name == "STOP":
            break
        elif usr_file_name != "dna7.fsa":
            print("Can only work on dna7.fsa")
        else:
            pass

        some_dna_list = get_dna_list(usr_file_name)
        converted_list = []
        for one_dna_string in some_dna_list:
            all_pairs = get_pairs(one_dna_string)
    # first you have to create the dictionary.
            conv_dict = get_dict(all_pairs)
    # will need to consider even or odd lengths of the dna string when you
    # actually inplement the dictionary to convert.
            converted_one_string = return_conv(conv_dict, one_dna_string)
            converted_list.append(converted_one_string)
    
        print(converted_list)



def get_dna_list(file_name):
    with open(f"{file_name}", "r") as infile:
        dna_strings = ""
        for line in infile:
            line = line.strip()
            if line.startswith(">"):
                line = ">"
            else: 
                pass
            dna_strings += line
    dna_strings_as_list = dna_strings.split(">")
    dna_strings_as_list.pop(0)    # removes the empty ">" where we did the first split
    return dna_strings_as_list

def get_pairs(whole_dna_string):    # correct
    nucleotides = list(set(whole_dna_string))
    list1, list2, comb_list = [], [], []

    for _ in range(len(nucleotides)):
        for nucleotide in nucleotides:
            list1.append(nucleotide)
    for nucleotide in nucleotides:
        for _ in range(len(nucleotides)):
            list2.append(nucleotide)
    for position in range(max(len(list1), len(list2))):
            pair = list1[position] + list2[position]
            comb_list.append(pair)

    # returns left over length
    
    return(comb_list)


# given the list of all possible pair combinations
# returns a dictionary where the pairs are the keys, the keys are the asciis.
def get_dict(comb_list):    # correct
    comb_dict = {}
    ascii_values = ["E", "I", "L", "O", "P", "Q", "V", "W", "X", "Y", "Z", "a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p"]
    
    for position in range(len(comb_list)):
        comb_dict[comb_list[position]] = ascii_values[position]
    
    return (comb_dict)

# given the conversion dictionary and the whole dna string.
# returns a string only containing the converted value
# takes into account situations where there are an odd number of nucleotides.
def return_conv(comb_dict, whole_dna_string):
    return_string = ""
    pairs = []
    for position in range(0, len(whole_dna_string), 2):    # specify two to step by two
        #sanity check, last one should be left out, yes
        pair = whole_dna_string[position:position + 2]
        pairs.append(pair)
    if len(whole_dna_string) % 2 == 0:
        for pair in pairs:
            return_string += comb_dict[pair]
    elif len(whole_dna_string) % 2 == 1:
        for pair in pairs[:-1]:    # all value up until the final one.
            return_string += comb_dict[pair]
        return_string += pairs[-1]    # then we add the odd one out.

    return return_string


main()
```
