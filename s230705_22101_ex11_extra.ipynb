A k-mer is a piece of sequence k units long. It can be both protein and DNA sequence. As an example I have this DNA sequence: GACTAC. It contains the following 4-mers: GACT, ACTA, CTAC.

1\. From a genbank file like data1-4.gb extract the DNA sequence (been there - done that). Now insert all 5-mers in the DNA sequence into a set. Compute the following numbers: how many did you insert in the set, how many are in the set, how many different 5-mers could there possibly be. Display. It seems like the two first numbers are the same, but this is not guaranteed to be true, why?

```python
# pseudo code
"""
prompt user for a file within data1-4.gb
with open the file
    use stateful parsing to extract the DNA sequence into one string..

trying to write a function for 4-mers and expand it to apply to 5-mer
    will make use of slicing and a for loop.

number of possibl different 5-mers can be calculated by editing my fuction to calculate number of possible codons.
now instead of making a string of len = 3 for each codon using different combinations of "ATCG",
I will make a strng of len = 5 for each codon using different combinations of "ATCG".

"""
def main():
    while True:
        usr_file = input("Which gene-bank file do you want to extract the DNA sequence from?\nChoose one from data1-4.gb (STOP to quit): ")
        if usr_file == "STOP":
            break
        else:
            usr_seq = get_dna_seq(usr_file)
            usr_n = int(input("We will give you a lot of n-mers, specify your n: "))
            (n_insertions, usr_nmer_set) = n_mer_set(usr_seq, usr_n)
            print(f"There were {n_insertions} 5-mer insertions to the set.")
            print(f"The set contains {len(usr_nmer_set)} 5-mers in total.")

            # how many different n-mers can there possibly be?
            # modify the codon generator to make all uniqe 5-mers
            # includes 5-mers where all 5 nucleotides are the same, i.e. "AAAAA"
            all_5mers_set = make_all_5mers()
            print(f"There are {len(all_5mers_set)} possible unique 5-mers in total.")
            # check out which unique 5mers were not part of the n_mer_set from a given file:
            print(f"Out of all possible unique 5-mers, the 5-mers from {usr_file} included all but the following: ")
            print(sorted(all_5mers_set.difference(usr_nmer_set)))


# given seq string and n_mer number
# returns possible n_mers
def get_dna_seq(file):
    with open(f"{file}", "r") as infile:
        (dna_seq, dna_flag) = ("", False)
        for line in infile:
            if line[:-1] == "//":
                dna_flag = False
            if dna_flag:
                dna_seq += "".join(line.strip().split()[1:])
            if "ORIGIN" in line:
                dna_flag = True
    if dna_flag:
        raise ValueError("Format is not right. Cannot trust result.")
    if len(dna_seq) == 0:
        raise ValueError("No data found. Cannot trust result.")
    # return dna_seq instead of print dna_seq here :)
    # so that in can be used as input for the n_mer function in main.
    return(dna_seq)


# This one addes each n_mer into a set, and counts the number of times of insertions.
# It does not print out each n_mer as they are made.
def n_mer_set(seq, n):
    new_set = set()
    # counts "how many n-mers" I inserted into the set
    insertion_counts = 0
    # this is to prevent len(seq) from being interpreted as a string of len = 3. 
    len_seq = len(f"{seq}")

    # a list of all the possible starting positions for the n-mer
    # +1 to adjust for the index, try with a toy example if you forget:
    start_positions = list(range(0,(len_seq - n) + 1))    
    for start_position in start_positions:    # this is the starting value
        end = start_position + n  # notice this is the correct index format for slicing.
        new_set.add(seq[start_position:end])
        insertion_counts += 1
    
    return(insertion_counts, new_set)


# code adopted from https://www.geeksforgeeks.org/python-program-to-print-all-possible-combinations-from-the-three-digits/
# to make all 5-mers:
def make_all_5mers():
    L = "atcg"    # remember to set this to lower case
    fivemers_set = set()
    len_nucleotides = len(L)
    for i in range(len_nucleotides):
        for j in range(len_nucleotides):
            for k in range(len_nucleotides):
                for l in range(len_nucleotides):
                    for m in range(len_nucleotides):
                        fivemer = L[i]+L[j]+L[k]+L[l]+L[m]
                        fivemers_set.add(fivemer)
    # sanity check
    return(fivemers_set)
    
    

main()
```

3\. Calculate the overlap of 5-mers between any two of the data1-4.gb files. Just ask for two filenames and calculate for these.

```python
# pseudo code:
"""
prompt user for two file names

extract the dna sequence from the file

run each file's dna seq through a function that will return a set of 5-mers

use the set method .intersection to find out the 5-mers that do overlap, and how many there are.
"""

def main():
# prompt user for two filenames:
    while True:
        # use .spit() to save the two filename inputs as two separate variables
        try:
            usr_file1, usr_file2 = input("Please enter two file names from data1-4.gb, separated by space (STOP NOW to exit): ").split()
        except ValueError as error:
            print("Please enter two filenames.")
        except FileNotFoundError:
            print("Please enter two filenames from data1-4.gb")
        else:
            two_usr_files = [usr_file1, usr_file2]
            if two_usr_files == "STOP NOW".split():
                break
    
        # extract dna seq of both files
        two_usr_file_seqs = []
        for usr_file in two_usr_files:
            # make sure get_file_seq will return a str.
            two_usr_file_seqs.append(get_dna_seq(usr_file))
        
        # generate 5-mers of both files
        usr_n = int(input("Please enter the value of n for your n-mer: "))
        two_usr_nmer_sets = []
        for usr_seq in two_usr_file_seqs:
            two_usr_nmer_sets.append(n_mer_set(usr_seq, usr_n))
        
        # sanity check to make sure we have sets within a list:
        # print(type(two_usr_nmers_sets[0]))
        n_mer_set_intersection = two_usr_nmer_sets[0].intersection(two_usr_nmer_sets[1])
        print(f"{usr_file1} and {usr_file2} contain {len(n_mer_set_intersection)} {usr_n}-mer overlaps")
    # before moving on to index the list above and find intersection.

# modified from L11_ex1's n_mer_set.
# takes DNA seq strings and n_mer's n as input
# only returns a set of n_mers
# does not return insertion coutns or lenth count.
def n_mer_set(seq, n):
    new_set = set()
    # this variable prevents seq from being interpreted as a string of len 3.
    len_seq = len(f"{seq}")

    start_positions = list(range(0, (len_seq - n)+1))
    for start_position in start_positions:
        end = start_position + n
        new_set.add(seq[start_position:end])
    
    return(new_set)


# input genebank format filename
# outputs one string of dna
def get_dna_seq(file):
    with open(f"{file}", "r") as infile:
        (dna_seq, dna_flag) = ("", False)
        for line in infile:
            if line[:-1] == "//":
                dna_flag = False
            if dna_flag:
                dna_seq += "".join(line.strip().split()[1:])
            if "ORIGIN" in line:
                dna_flag = True
    if dna_flag:
        raise ValueError("Format is not right. Cannot trust result.")
    if len(dna_seq) == 0:
        raise ValueError("No data found. Cannot trust result.")
    # return dna_seq instead of print dna_seq here.
    # so that it can easily be used as an input for a different function.
    return(dna_seq)
    




main()
```

3\. Make a program that asks for a number (integer) then reads the dna7.fsa file (which contains insulin-like genes) and saves the entry selected by the number in selected.fsa and the rest of the entries into the file rest.fsa.

```python
# pseudo code
"""
main function:
    prompt user for a number
        convert it to an int.

    prompt user for a file name (can only be dna7.fsa in this cse)
    use fsa_entry_to_list function on file.
    use list_to_files on fsa_entry_to_list and usr's selected number.
        
    function fsa_entry_to_list: input file name, outputs list of entries.
        read the dna7.fsa file
    put each entry into a list
    using a list because lists are ordered

    function list_to_files: input list of entries AND selected number, output two files.
        access the correct index on the list to save the entry selected by the usr into selected.fsa,
        can do this by returning the output of .pop(correct index)

        save the list after doing .pop() into file rest.fsa

return the main function
"""

def main():
    # the filename will only be dna7.fsa in this case.
    while True:
        usr_filename = input(f"Please enter the .fsa filename you want to look at (i.e. dna7.fsa): ")
        if usr_filename == "dna7.fsa":
            usr_number = int(input(f"Please enter the entry number you want to select from {usr_filename}: "))
            usr_fsa_as_list = fsa_entry_to_list(usr_filename)
            list_to_files(usr_fsa_as_list, usr_number)
            # break here so we leave once we get wehat we need :)
            break
        else:
            print("Sorry, we only have dna7.fsa to work on, pleace try again and enter dna7.fsa")





def list_to_files(entries_as_list, entry_number):
    # make sure to use the correct index number:
    index_number = entry_number - 1
    # after pop, we have an list that was changed in-place.
    selected_entry = entries_as_list.pop(index_number)
    rest_entries_list = entries_as_list
    with open("selected_entry.fsa", "w") as selected_outfile:
            # output file displayed in a pretty way because of correct when reading the file earlier.
            # i.e. by not doing line = line.strip() when reading the file.
            selected_outfile.write(selected_entry)
    with open("rest_entries.fsa", "w") as rest_outfile:
        for entry in entries_as_list:
            rest_outfile.write(entry) 



def fsa_entry_to_list(file):
    seq = ""
    with open(f"{file}", "r") as infile:
        for line in infile:
            # rember not to do line = line.strip() here
            # it will help to keep your output file format pretty :)
            seq += line
    # thinking to use ">" split, checking the output.
    # use replace(">", "@>").split("@") to retain > but also achieve correct splitting.
    seq_list = seq.replace(">","@>").split("@")
    # pop(0) removes the leading "" in the list, which we don't need.
    seq_list.pop(0)
    return(seq_list)




main()
```
