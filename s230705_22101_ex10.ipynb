5\. Improve exercise 5.8 using all you have learned. The program shall now take a DNA FASTA file (asking interactively for it), and reverse and complement all entries in the file. There can be more than one entry, study dna7.fsa. Hint: Use substitution or translation for complementing the DNA. I will point out that the reading and writing of fasta files with many entries is a regular occurrence in bioinformatics (and exam), so be sure to get it right. Many people mistakenly believe that they should use a form of stateful parsing with a flag for this - doing so confuses the issue, so abstain from that.

```python
# pseudo code
"""
Summary: Keep the first line and reverse complement the sequence

can utilise translate string for complementing the DNA.
make a main function first
    prompts user for DNA FASTA file name and saves that into a variable.
    first with open an outfile, where we will write the reverse_comp into
        second with open an infile, i.e. the file specified by the user, from which we read the information
            for line in infile
                strip the lines of unneeded white space
                if the line starts with ">":
                    we write it to the outfile directly.
                else:
                    we run the reverse_complement function on the line. The output of this function is written into the outfile.

what the reverse_complement function does:
    takes in one_line
    creates a translationTable for DNA nucleotide complementarity.
    uses the .translate() method to translate the string in the given line.
    converts this string into a list
    uses .rever() to reverse this list inplace
    uses join to save this list as a string variable, rev_comp_one_line
    this small function returns the string variable, rev_comp_one_line

The output file has each of the "headerlines" that begin with ">" intact,
followed by DNA nucleotides presented in the same format as before, only that now they are the reverse complements of the original :)
"""
# code
def main():
    usr_file = input("Please enter a DNA FASTA file name, such as dna.fsa, dna7.fsa: ")
    with open(f"{usr_file}".split(".")[0]+"_reverse_comp.fsa", "w") as outfile:
        with open(f"{usr_file}", "r") as infile:
            for usr_line in infile:
                usr_line = usr_line.strip()
                if usr_line.startswith(">"):
                    outfile.write(usr_line)
                    outfile.write("\n")
                else:
                    outfile.write(rev_complement(usr_line))
                    outfile.write("\n")

def rev_complement(one_line):
    translationTable = str.maketrans("ATCG", "TAGC")
    complement_one_line = one_line.translate(translationTable)
    comp_list = list(complement_one_line)
    comp_list.reverse()
    rev_comp_one_line = ''.join(comp_list[:])
    return rev_comp_one_line


main()
```
