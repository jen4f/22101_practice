1\. Ask for 2 input file names (ex1.acc & ex1.dat are good) and an output file name. Copy (read and write) the input files into the output file, one after the other.

```python
# prompt user for input file names
file1 = input("Enter the first file name: ")
file2 = input("Enter the second file name: ")
output = input("Enter the name you want for the output file: ")

# read the first file
outfile = open(f"{output}", "w")
with open(f"{file1}", "r") as infile1:
    for line in infile1:
        outfile.write(line)
with open(f"{file2}", "r") as infile2:
    for line in infile2:
        outfile.write(line)
outfile.close()
```

2\. Ask for 2 input file names (ex1.acc & ex1.dat are good) and an output file name. Now copy\/print the lines in the input files into the output file so that the first line of each input file are merged together with a tab and becomes the first line of the output. Continue that way with all the lines in the input file. A mental picture of what is supposed to happen is this: Imagine the 2 input files to be two pieces of paper with lines. In the exercise before the papers were put one after the other in the output. In this exercise the papers are put next to each other (sideways) in the output.


```python
# prompt user for input file names
file1 = input("Enter the first file name: ")
file2 = input("Enter the second file name: ")
output = input("Enter the name you want for the output file: ")

# readlines() returns a list, each line has a \n at the end.
# example item from ex1.dat in a list: '0.7862\t-0.7197\t2.6020\n'
# example item from ex1.acc in a list: 'RPB1_HUMAN\tX74874.CDS.1\n'
with open(f"{file1}", "r") as infile1:
    part1 = infile1.readlines()
with open(f"{file2}", "r") as infile2:
    part2 = infile2.readlines()

# check if the two input files have the same number of items
print(f"The two input files have the same number of items: {len(part1) == len(part2)}")

# initiate an emptry string to write contents of output file
combined = ""
# implement a while loop to weave the two lists together.
i = 0
while i != len(part1):
    combined += part1[i].strip() + "\t" + part2[i]    #part2[i] will make sure there is a \n.
    i += 1
# sanity check, the outfile will contain this.
print(combined)

with open(f"{output}", "w") as outfile:
    outfile.write(combined)
```

3\. Read the mixedlines.txt and count how many long lines there are in the file. A long line has more than 30 chars.


```python
# initiate counter variable
counter = 0
with open("mixedlines.txt", "r") as infile:
    for line in infile:
        line = line.strip()
        if len(line) > 30:
            counter += 1
        else:
            pass
print(f"mixedlines.txt has {counter} long lines.")
```

4\. In mixedlines.txt how many occurrences for "rna" can you find?

```python
import os

# initiate counter varialble
counter = 0
match_text= os.popen("grep rna mixedlines.txt").read()

# code below was written based on this lecture's ppt, 22101_06-InputOutput, slide 12.
for i in range(len(match_text)):
    if match_text[i:i+3] == "rna":
        counter += 1
    else:
        pass
print(f"'rna' occurs in these parts of the text:\n{match_text.strip()}.")
print(f"\nthe word string 'rna' occurs {counter} times.")
```

5\. In mixedlines.txt how many lines contain "rna"?

```python
import os
# occurences of rna in the file, i.e. rna occurs consecutively.
match_text= os.popen("grep rna mixedlines.txt").read()
match_number = os.popen("grep -c rna mixedlines.txt").read()

print(f"'rna' occurs in these parts of the text:\n{match_text.strip()}.")
print(f"\n'rna' occurs on {match_number.strip()} lines.")
```
