3\. This exercise requires that you did the last two practice exercises in Simple Pattern Matching. In the data1-4.gb files count who many times the different codons in the coding sequence occurs. Display.

```python
# pseudo code
"""
def main()
for the list of files, do below for each file:
how I would do it for one file:
    input one file into get_cds_positions(), which will return one list of cds positions
    input one file into get_seq(), which will return one list of the dna seq.
    input the reutrn values of the above, i.e. the list of cds positions, and the dna seq to get_cds(),
        which will return a list of actual cds, with the associated dna.
    input one list of cds into codon_freq(), will return a dictionary, which counts the occurence of different codons.

what does get_cds_positions() do?
use simple pattern matching to extrace the coding positions from a file into one list.

what does get_seq() do?
use simple pattern matching to extract the cds

what does get_cds() do?
given a list of cds positions, and a string of a dna sequence
returns a list of dna sequences, with each string being the coding sequence.


what does codon_freq() do?
initiate an empty codon_freq_dict
makes a list that is equal to the three stop codon strings.
loop through the codon list, making sure to slice it correctly:
    if these three are not in the list(codon_freq_dict.key()):
        codon_freq_dict[these three] = 1    # ininitate count.
    elif these three are equal to the stop codons:
        break
    else these three are already in the list(codon_freq_dict.key()):
        codon_freq_dict[these three] += 1
at the end of the loop, return a dictionary codon_freq_dict.
"""
# code
def main():
    usr_positions = get_cds_positions("data2.gb")
    usr_dna_seq = get_dna_seq("data2.gb")
    usr_dna_cds = get_cds(usr_positions, usr_dna_seq)
    print(codon_counts(usr_dna_cds))



# recieves a list of aa sequences used for coding
# returns the counts of occurences of different codons.
def codon_counts(dna_cds):
    # inidiate empty counter dict:
    codon_counts = {}
    for item in dna_cds:
        for position in range(0,len(item),3):
            codon = item[position:position+3]
            # adjust to the new correct start of the reading frame.
            if codon in codon_counts:
                codon_counts[codon] += 1
            else:
                codon_counts[codon] = 1
    return codon_counts


def get_cds(positions, seq):
    cds = []
    for position in positions:
        start_position = int(position.split("..")[0]) - 1 # use -1 to account for zero indexing.
        end_position = int(position.split("..")[1])
        cds.append(seq[start_position:end_position])
    return cds



def get_dna_seq(file):
    with open(f"{file}", "r") as infile:
        (seq, seq_flag) = ("", False)
        for line in infile:
            if line.startswith("//"): # Red Flag
                seq_flag = False
            if seq_flag:
                # [1:] will exclude the leadning numbers.
                seq += "".join(line.lstrip().split()[1:])
            if line.startswith("ORIGIN"): # Green Flag
                seq_flag = True
        if seq_flag:
            raise ValueError("Format incorrect, cannot trust result.")
        if len(seq) == 0:
            raise ValueError("No data found, cannot trust result.")
    return seq




# takes any genebank file (.gb) as input
# returns a list of coding sequences (cds), where each item a string of the coding region range.
def get_cds_positions(file):
    with open(f"{file}", "r") as infile:
        (cds_p, cds_flag) = ("", False)
        for line in infile:
            if line.startswith("     CDS             join"):   # Green Flag
                cds_flag = True
            if cds_flag:
                #.strip(")") to get rid of the extra ")" at the end.
                cds_p += line.strip("     CDS").replace("join(", "").replace(")", "").replace('/gene="COL1A2"','').replace("\n","")
            if line.strip().startswith("/"):    # Red Flag
                cds_flag = False
        if cds_flag:
            raise ValueError("Format incorrect, cannot trust result.")
        if len(cds_p) == 0 :
            raise ValueError("No data found, cannot trust result.")
    # cds is correctly retrieved
    cds_p = cds_p.split(",")
    return cds_p



main()
```
