1\. Read two files with numbers into two lists. Files ex1_1.dat, ex1_2.dat & ex1_3.dat are good to use for this. Now add the numbers together, like the first number in the first file with the first number in the second file, and so forth, i.e. add the column in the file row-wise. You could also consider the numbers in a file to be a vector, and then you add two vectors together. Save the output (a list of numbers, one number per line) in a new file - you decide the name.


```python
# pseudo code
"""
use with open two seperate times to read into two lists

find the longest length of the two, default to first list
longest_length = max(len(first_list), len(second_list))
creat an empty string full of zeros with the same length as the longest list

for position in range(len(first_list)):
    empty_list[position] + first_list[position]
for position in range(len(second_list)):
    empty_list[position] + second_list[position]
return empty_list

write list in to new file, one line per list item.
"""
# code

first_file = input("Enter the name of one file from the following:\nex1_1.dat, ex1_2.dat, ex1_3.dat\nFirst file name: ")
second_file = input("\nEnter the name of one file from the following:\nex1_1.dat, ex1_2.dat, ex1_3.dat\nSecond file name: ")

first_list = []
second_list = []

with open(f"{first_file}", "r") as first_infile:
    for line in first_infile:
        first_list.append(float(line.strip()))
with open(f"{second_file}", "r") as second_infile:
    for line in second_infile:
        second_list.append(float(line.strip()))

longest_length = max(len(first_list), len(second_list))
empty_list = [0] * longest_length
for position in range(len(first_list)):
    empty_list[position] += first_list[position]
for position in range(len(second_list)):
    empty_list[position] += second_list[position]

# edit list so all values have the same number of decimal places
for position in range(len(empty_list)):
    empty_list[position] = round(empty_list[position], 4)
print(empty_list)

# sanity check 
if (len(empty_list) == longest_length):
    print("Result has the correct number of items :-)")
else:
    print("Result does not have the equal number of items, something is wrong :<")

# save the output (a list of numbers, one number per line) in a new file.
sum_filename = f"{first_file.split('.')[0]}plus_{second_file.split('.')[0]}.{first_file.split('.')[1]}"
with open(f"{sum_filename}", "w") as outfile:
    for item in empty_list:
        outfile.write(str(item)+"\n")
```

```python
# code
# same as in ex1, but with some minor chances, see in-code comments.
# any words or empty lines present will be processed as zero

first_file = input("Enter the name of one file from the following:\ntext_1.dat, text_2.dat\nFirst file name: ")
second_file = input("\nEnter the name of one file from the following:\ntext_1.dat, text_2.dat\nSecond file name: ")

first_list = []
second_list = []

with open(f"{first_file}", "r") as first_infile:
    for line in first_infile:
        if len(line.strip()) == 0:    # check to see if the line is empty
            first_list.append(0)    # pretend there is a zero on the line.
        elif line.strip().isalnum():    # checks to exclude any words, or letters mixed with numbers
            first_list.append(0)    # pretend there is a zero on the line.
        else:
            first_list.append(float(line.strip()))

with open(f"{second_file}", "r") as second_infile:
    for line in second_infile:
        if len(line.strip()) == 0:    # check to see if the line is empty
            second_list.append(0)    # pretend there is a zero on the line.
        elif line.strip().isalnum():    # check to see if the line is a word with no numbers.
            second_list.append(0)    # pretend there is a zero on the line.
        else:
            second_list.append(float(line.strip()))

longest_length = max(len(first_list), len(second_list))
empty_list = [0] * longest_length
for position in range(len(first_list)):
    empty_list[position] += first_list[position]
for position in range(len(second_list)):
    empty_list[position] += second_list[position]

# edit list so all values have the same number of decimal places
for position in range(len(empty_list)):
    empty_list[position] = ("{0:.4f}".format(empty_list[position]))
print(empty_list)
```
